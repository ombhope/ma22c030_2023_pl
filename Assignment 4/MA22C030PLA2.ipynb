```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin, cos
```

```python
# Define a piecewise function
def my_piecewise(x):
    if x >= 0:
        return cos(x)
    else:
        return x * sin(x * 2)
```

```python
# Create an array of uniformly spaced points in the interval [-5, 5]
x = np.linspace(-5, 5, 1000001)

# Compute the values of the piecewise function
y = np.array([my_piecewise(i) for i in x])
```

```python
# Plot the function
plt.plot(x, y, '-')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Piecewise Function')
plt.grid(True)
plt.show()
```
