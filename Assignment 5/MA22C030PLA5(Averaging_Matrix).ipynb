```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_alternating_vec(n):
    x = np.zeros(n)
    for i in range(n):
        x[i] = (-1) ** i
    return x
```

```python
def compute_and_plot_derivative(n, k, f_x):
    y = np.zeros((n, n))
    for i in range(1, k + 1):
        diag = np.ones(n - i)
        y += np.diag(diag, i) + np.diag(diag, -i)

    avg = np.sum(y, axis=1)
    for i in range(len(avg)):
        y[i, :] = y[i, :] / avg[i]

    y1 = np.matmul(y, f_x)

    plt.plot(x, y1, label=f'n={n}, k={k}')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()
```

```python
# Generate the input data
x = np.linspace(0, 10, 1000)
f_x = x**2 + 2 * np.sin(10 * x)
```

```python
# Compute and plot derivatives for different values of k
plt.figure(figsize=(12, 6))
```

```python
k_values = [97, 98, 99, 100, 101, 102, 103, 300, 25]
for k in k_values:
    compute_and_plot_derivative(1000, k, f_x)
```

```python
plt.plot(x, f_x, label='Actual Graph')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
