```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
```

```python
# Create the difference matrix
dif_mat = np.diag(np.ones(n)) - np.diag(np.ones(n-1), k=-1)
```

```python
x = np.linspace(0, 2*np.pi, n)
y = np.sin(x)
y1 = np.dot(dif_mat, y)
y1 /= (2*np.pi / n)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x[1:], y1[1:], label='cos(x)')
plt.plot(x, y, label='sin(x)')
plt.legend(loc='best')
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, n)
y = x**2 + 2*np.sin(10*x)
y1 = np.dot(dif_mat, y)
y1 /= (2*np.pi / n)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x[1:], y1[1:], label='derivative of F(x)')
plt.plot(x, y, label='F(x)')
plt.legend(loc='best')
plt.show()
```

```python
# Efficient Approach
def get_diff_mat(n):
    diff_mat_eff = np.diag(np.ones(n), k=0) - np.diag(np.ones(n-1), k=-1)
    return diff_mat_eff
```

```python
diff_mat_eff = get_diff_mat(n)
```

```python
diff_mat_eff.shape
```

```python
x_new = x
y_new = x_new**2 + 2*np.sin(10*x_new)
```

```python
diff_mat_eff.shape
```

```python
y1_new = np.dot(diff_mat_eff, y_new)
y1_new /= (2*np.pi / n)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_new, y1_new, label='derivative of F(x)')
plt.plot(x, y, label='F(x)')
plt.legend(loc='best')
plt.show()
```
