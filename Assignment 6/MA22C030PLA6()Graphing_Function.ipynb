```python
import matplotlib.pyplot as plt   # Import the matplotlib library and rename it as plt.
import numpy as np    # Import the numpy library and rename it as np.
```

```python
%matplotlib inline

# Generate an array of x values in the range [0, 10]
x_values = np.linspace(0, 10, 1000)  # You can adjust the number of points for a smoother or coarser curve

# Calculate the corresponding y values for the sine function
y_values = np.sin(x_values)
```

```python
# Create a plot of the function
plt.plot(x_values, y_values)
plt.xlabel('x-axis')
plt.ylabel('y-axis (sin(x))')
plt.title('Graph of the Sine Function')
plt.grid(True)
plt.show()
```
