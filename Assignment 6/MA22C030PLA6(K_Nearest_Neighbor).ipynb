```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def k_nearest(points, k):
    points = np.array(points)
    num_points = points.shape[0]
    k_nearest_neighbors = []

    for i in range(num_points):
        distances = np.linalg.norm(points - points[i], axis=1)
        sorted_indices = np.argsort(distances)
        k_nearest = [points[j] for j in sorted_indices[1:k + 1]]  # Exclude the point itself
        k_nearest_neighbors.append(k_nearest)

    return k_nearest_neighbors
```

```python
points = [[0, 0], [0.2, 0.22], [0.1, -0.1], [1, 1], [1.1, 0.9], [0.8, 0.9], [1.1, 0.63],
     [0.58, -0.1], [0.63, 0.1], [0.67, -0.3], [0.8, -0.23], [0.8, 0.6]]
k_value = 3
```

```python
k_nearest_neighbors = k_nearest(points, k_value)
```

```python
def plot_direction_arrows(points, neighbors):
    for i, point in enumerate(points):
        for neighbor in neighbors[i]:
            plt.annotate("", xy=neighbor, xytext=point, arrowprops={"arrowstyle": '->', 'color': 'magenta'})





# def plot_direction_arrows(points, neighbors):
#     for i, point in enumerate(points):
#         for neighbor in neighbors[i]:
#             plt.annotate("", xy=neighbor, xytext=point, arrowprops={})

plt.figure(figsize=(10, 8))

# Plot the points
x, y = zip(*points)
plt.scatter(x, y, c='blue', marker='o', label='Points')
plot_direction_arrows(points, k_nearest_neighbors)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('K-Nearest Neighbors')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
