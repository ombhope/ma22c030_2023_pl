```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Set different URL for COVID-19 data
url_new = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'

```

```python
# Load data from the new URL
data_new = pd.read_csv(url_new)
data_new.head()

```

```python

if 'data_new' in locals() and 'Country/Region' in data_new.columns:
    countries = ['Italy', 'Spain', 'Germany']
    country_data = data_new[data_new['Country/Region'].isin(countries)]
    num_records_new = len(country_data)

    print(f"Number of records in Italy, Spain, and Germany: {num_records_new}")
else:
    print("The 'data_new' DataFrame or 'Country/Region' column does not exist.")

# Drop unnecessary columns in the DataFrame
columns_to_drop_new = ['Lat', 'Long', 'Province/State']
country_data = country_data.drop(columns=[col for col in columns_to_drop_new if col in country_data.columns])
country_data.head()

country_data = country_data.groupby('Country/Region').sum().T

fig, ax = plt.subplots()
fig.set_facecolor('lightgray')
country_data['Italy'].plot(color="black")
country_data['Spain'].plot(color="blue")
country_data['Germany'].plot(color="green")
plt.legend()
plt.axis()
plt.xlabel("Date")
plt.ylabel("Confirmed COVID-19 cases")

plt.show()
```
