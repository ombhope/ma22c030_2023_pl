```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read dataset
df = pd.read_csv('/content/co2_mm_mlo.csv')
```

```python
df.head
```

```python
# Convert df to array
df_as_array = np.array(df)
df_as_array.shape
```

```python
df_as_array
```

```python
x = df_as_array[:,2]
y = df_as_array[:,3]
```

```python
x_modified = np.linspace(1958.2877 ,2023.4583, num=783)
y_modified = np.interp(x_modified, x, y)
```

```python
# Plot the interpolant
plt.plot(x_modified, y_modified, '-', label='Interpolant')
plt.legend(loc='best') # Best fit
plt.show()
```
